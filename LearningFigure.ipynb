```python
from kalman_gst import dirichlet_covar
```

```python
p_obs = [0.3, 0.7]
p_pred = [0.4, 0.6]
counts = 2
covar = dirichlet_covar(counts*p_pred)
```

```python
fig, axs = plt.subplots(2, 1, figsize=(4, 10))
axs[0].set_title("Observation")
axs[0].set_yticklabels([])
axs[0].set_xticklabels([])
axs[1].set_xticklabels([])

axs[0].set_ylim([0, 1])
axs[1].set_ylim([0, 1])
axs[0].set_ylabel('Outcome 0 (frequency)')
axs[1].set_ylabel('Outcome 1 (frequency)')
axs[1].set_xlabel('Probability density')

y_range = np.linspace(0, 1, 100)
axs[0].plot((0, 10), (p_obs[0], p_obs[0]), c='dodgerblue')
axs[1].plot((0, 10), (p_obs[1], p_obs[1]), c='dodgerblue')
axs[0].plot(norm.pdf(y_range, p_pred[0], covar[0, 0]), y_range, c='orange')
axs[1].plot(norm.pdf(y_range, p_pred[1], covar[1, 1]), y_range, c='orange')
axs[0].plot(norm.pdf(y_range, p_pred[0], covar[0, 0]), y_range, c='orange')
axs[1].plot(norm.pdf(y_range, p_pred[1], covar[1, 1]), y_range, c='orange')
```

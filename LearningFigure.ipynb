```python
fig, axs = plt.subplots(2, 2,  figsize=(10,10))

xax = axs[0, 0].get_xaxis()
xax = xax.set_visible(False)
xax = axs[0, 1].get_xaxis()
xax = xax.set_visible(False)
# xax = axs[1, 0].get_xaxis()
# xax = xax.set_visible(False)
xax = axs[1, 1].get_xaxis()
xax = xax.set_visible(False)
#xax = axs[0, 0].get_yaxis()
#xax = xax.set_visible(False)
xax = axs[0, 1].get_yaxis()
xax = xax.set_visible(False)
axs[1, 0].axes.yaxis.set_ticklabels([])
axs[0, 0].axes.yaxis.set_ticklabels([])

xax = axs[1, 1].get_yaxis()
xax = xax.set_visible(False)

axs[0, 0].set_xlim([-2, 3])
axs[0, 1].set_xlim([-2, 3])

x_estimate = 1.4

x_range = np.linspace(-np.pi, np.pi, 1000)
h_short = (np.cos(x_range)+1)/2
h_long = (np.cos(5*x_range) + 1)/2
y_likelihood_short = binom.pmf(50, 100, h_short)
axs[0, 0].plot(x_range, 10*y_likelihood_short, c='dodgerblue')
#axs[0, 0].plot(x_range, 10*y_likelihood_short_linearized, c='red', linestyle='dotted', linewidth=3)
# axs[0, 1].plot(x_range, 10*y_likelihood_long_linearized, c='red', linestyle='dotted', linewidth=3)
#axs[0].plot(x_range, (np.cos(x_range)+1)/2)

wide_prior = norm.pdf(x_range, x_estimate, 0.75)
narrow_prior = norm.pdf(x_range, x_estimate, 0.1)


axs[0, 0].plot(x_range, wide_prior, c='green',  linestyle='dashed')
axs[0, 0].plot(x_range, 0.3*narrow_prior, c='orange',  linestyle='dashed')

y_likelihood_long = binom.pmf(50, 100, h_long)

axs[0, 1].plot(x_range, 10*y_likelihood_long)

axs[0, 1].plot(x_range, wide_prior, c='green', linestyle='dashed')
axs[0, 1].plot(x_range, 0.3*narrow_prior, c='orange', linestyle='dashed')

axs[0, 0].plot((np.pi/2, np.pi/2), (-.05, 0.1), c='black')

axs[0, 1].plot((np.pi/2, np.pi/2), (-.05, 0.1), c='black')
axs[1, 1].plot((np.pi/2, np.pi/2), (-.005, 0.01), c='black')

h_linearized_short = (1+np.cos(x_estimate))/2 - (np.sin(x_estimate)/2)*(x_range-x_estimate)
h_linearized_long = (1+np.cos(5*x_estimate))/2 - 5*(np.sin(x_estimate)/2)*(x_range-x_estimate)
y_likelihood_short_linearized = binom.pmf(50, 100, h_linearized_short)
y_likelihood_long_linearized = binom.pmf(50, 100, h_linearized_long)





prod_short_narrow = y_likelihood_short*narrow_prior/(sum(y_likelihood_short*narrow_prior))
prod_short_wide = y_likelihood_short*wide_prior/(sum(y_likelihood_short*wide_prior))

y_likelihood_short_linearized = np.nan_to_num(y_likelihood_short_linearized)
prod_short_narrow_linear = y_likelihood_short_linearized*narrow_prior/(sum(y_likelihood_short_linearized*narrow_prior))
prod_short_wide_linear = y_likelihood_short_linearized*wide_prior/(sum(y_likelihood_short_linearized*wide_prior))


axs[1, 0].plot(x_range, prod_short_wide, c='green')
axs[1, 0].plot(x_range, prod_short_narrow, c='orange')
#axs[1, 0].plot(x_range, prod_short_wide_linear, c='cyan', linestyle='dotted')
#axs[1, 0].plot(x_range, prod_short_narrow_linear, c='brown', linestyle='dotted')
axs[1, 0].plot((np.pi/2, np.pi/2), (-.0015, 0.003), c='black')


prod_long_narrow = y_likelihood_long*narrow_prior/(sum(y_likelihood_long*narrow_prior))
prod_long_wide = y_likelihood_long*wide_prior/(sum(y_likelihood_long*wide_prior))
y_likelihood_long_linearized = np.nan_to_num(y_likelihood_long_linearized)
prod_long_narrow_linear = y_likelihood_long_linearized*narrow_prior/(sum(y_likelihood_long_linearized*narrow_prior))
prod_long_wide_linear = y_likelihood_long_linearized*wide_prior/(sum(y_likelihood_long_linearized*wide_prior))


#axs[1, 1].plot(x_range, prod_long_narrow_linear, c='cyan')
#axs[1, 1].plot(x_range, prod_long_wide_linear, c='m')
axs[1, 1].plot(x_range, prod_long_narrow, c='orange')
axs[1, 1].plot(x_range, prod_long_wide, c='green')




axs[1, 0].set_xlim([-2, 3])
axs[1, 1].set_xlim([-2, 3])

# left, bottom, width, height = [0.77, 0.25, 0.13, 0.2]
# ax2 = fig.add_axes([left, bottom, width, height])
# xax = ax2.get_yaxis()
# xax = xax.set_visible(False)
# ax2.plot(x_range, prod_long_narrow)
# ax2.plot(x_range, prod_long_wide)
# ax2.plot(x_range, prod_long_narrow_linear, c='cyan', linestyle='dashed')
# ax2.plot(x_range, prod_long_wide_linear, c='m', linestyle='dashed')
# ax2.set_xlim([1.45, 1.65])
plt.subplots_adjust(wspace=0.05, hspace=0.05)

fontsize=18
axs[0, 0].legend(['Likelihood', 'Wide prior', 'Narrow prior', 'True value'], loc='upper left', fontsize=fontsize-4)
axs[1, 0].legend(['Wide posterior', 'Narrow posterior', 'True value'], loc='upper left', fontsize=fontsize-4)
axs[0, 0].set_title("Short circuit", fontsize=fontsize)
axs[0, 1].set_title("Long circuit", fontsize=fontsize)
axs[0, 0].set_ylabel("Prior", fontsize=fontsize)
axs[1, 0].set_ylabel("Posterior", fontsize=fontsize)
axs[1, 0].set_xlabel('Parameter value', fontsize=fontsize)
fig.savefig('Figures/Cartoon.png')
```
